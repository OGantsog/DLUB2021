```python
## Hello, Seaborn
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
fifa_filepath = 'Data/DataVis/fifa.csv'
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
fifa_data.head()
```

```python
## Plot the data

# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
```

```python
flight_filepath = "Data/DataVis/flight_delays.csv"
flight_data = pd.read_csv(flight_filepath, index_col="Month")
flight_data.head()
```

```python
iris_setosa = iris_data[iris_data.Species == 'Iris-setosa']
iris_versi = iris_data[iris_data.Species == 'Iris-versicolor']
iris_virgi = iris_data[iris_data.Species == 'Iris-virginica']
iris_virgi.tail()
```

```python
# Histograms for each species
sns.distplot(a=iris_setosa['Petal Length (cm)'], label="Iris-setosa", kde=False)
sns.distplot(a=iris_versi['Petal Length (cm)'], label="Iris-versicolor", kde=False)
sns.distplot(a=iris_virgi['Petal Length (cm)'], label="Iris-virginica", kde=False)

# Add title
plt.title("Histogram of Petal Lengths, by Species")

# Force legend to appear
plt.legend()
```

```python
# KDE plots for each species
sns.kdeplot(data=iris_setosa['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=iris_versi['Petal Length (cm)'], label="Iris-versicolor", shade=True)
sns.kdeplot(data=iris_virgi['Petal Length (cm)'], label="Iris-virginica", shade=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species")
```
